```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import date
import json
```

```python
Ticker_list = ["RACE", "TSLA", "FCAU", "GM", "HMC", "TM", "VLVLY"]
data = pd.DataFrame(columns=Ticker_list)
```

```python
data_query = yf.download(Ticker_list, start="2019-1-1", end="2020-2-13")
sorted_data = data_query['Adj Close']
final_data_adj_close_df = pd.DataFrame(sorted_data)
final_data_adj_close_df
```

```python
data_query_2 = yf.download(Ticker_list, start="2019-1-1", end="2020-2-13")
sorted_data_2 = data_query_2['Volume']
final_data_volume_df = pd.DataFrame(sorted_data_2)
final_data_volume_df
```

```python
volume_adjclose_merge = pd.merge(final_data_adj_close_df, final_data_volume_df, on = "Date", how = "outer")
volume_adjclose_merge
```

```python
volume_adjclose_merge = volume_adjclose_merge.rename(columns={'FCAU_x':"FCAU ADJ Close", 'GM_x':"GM ADJ Close", 'HMC_x':"HMC ADJ Close", 'RACE_x':"Race ADJ Close", 'TM_x':"TM ADJ Close", 'TSLA_x':"TSLA ADJ Close", 'VLVLY_x':"VLVLY ADJ Close",
       'FCAU_y':"FCAU Volume", 'GM_y':"GM Volume", 'HMC_y':"HMC Volume", 'RACE_y':"Race Volume", 'TM_y':"TM Volume", 'TSLA_y':"TSLA Volume", 'VLVLY_y':"VLVLY Volume"})
volume_adjclose_merge
```

```python
final_df = pd.DataFrame(volume_adjclose_merge)
final_df.head()
```

```python
final_df = final_df.dropna(how="any")
final_df
```

```python
final_df['FCAU ADJ Close'] = final_df['FCAU ADJ Close'].map("${:.2f}".format)
final_df['GM ADJ Close'] = final_df['GM ADJ Close'].map("${:.2f}".format)
final_df['HMC ADJ Close'] = final_df['HMC ADJ Close'].map("${:.2f}".format)
final_df['Race ADJ Close'] = final_df['Race ADJ Close'].map("${:.2f}".format)
final_df['TM ADJ Close'] = final_df['TM ADJ Close'].map("${:.2f}".format)
final_df['TSLA ADJ Close'] = final_df['TSLA ADJ Close'].map("${:.2f}".format)
final_df['VLVLY ADJ Close'] = final_df['VLVLY ADJ Close'].map("${:.2f}".format)
final_df['FCAU Volume'] = final_df['FCAU Volume'].map("{:,}".format)
final_df['GM Volume'] = final_df['GM Volume'].map("{:,}".format)
final_df['HMC Volume'] = final_df['HMC Volume'].map("{:,}".format)
final_df['Race Volume'] = final_df['Race Volume'].map("{:,}".format)
final_df['TM Volume'] = final_df['TM Volume'].map("{:,}".format)
final_df['TSLA Volume'] = final_df['TSLA Volume'].map("{:,}".format)
final_df['VLVLY Volume'] = final_df['VLVLY Volume'].map("{:,}".format)
final_df.head()
```
